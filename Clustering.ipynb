```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
data=pd.read_csv(r'cluster.csv')
data
```

```python
data.describe(include='all')
```

```python
data.isnull().sum()
```

```python
data['MINIMUM_PAYMENTS'].value_counts()
```

```python
data['MINIMUM_PAYMENTS']=data['MINIMUM_PAYMENTS'].fillna(value=data['MINIMUM_PAYMENTS'].mean())
```

```python
data['CREDIT_LIMIT']=data['CREDIT_LIMIT'].fillna(value=data['CREDIT_LIMIT'].mean())
```

```python
data.isnull().sum()
```

```python
x=data[['BALANCE', 'BALANCE_FREQUENCY', 'PURCHASES', 'ONEOFF_PURCHASES',
       'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE', 'PURCHASES_FREQUENCY',
       'ONEOFF_PURCHASES_FREQUENCY', 'PURCHASES_INSTALLMENTS_FREQUENCY',
       'CASH_ADVANCE_FREQUENCY', 'CASH_ADVANCE_TRX', 'PURCHASES_TRX',
       'CREDIT_LIMIT', 'PAYMENTS', 'MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT',
       'TENURE']]
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
wcss_score=[]
silhouette_scores=[]


for i in range(2, 8):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(x)
    wcss_iter=kmeans.inertia_
    wcss_score.append(wcss_iter)

    y_pred=kmeans.fit_predict(x)
    silhouette_score_avg=silhouette_score(x, y_pred)
    silhouette_scores.append(silhouette_score_avg)

    wcss_score_str='{:.10f}'.format(wcss_iter)
    print(f"Cluster {i}: WCSS={wcss_score_str}, Silhouette score={silhouette_score_avg}")
```

```python
num_ranges=range(2, 8)
plt.plot(num_ranges, wcss_score)
plt.xlabel('number of cluster')
plt.ylabel('wcss_score')
```

```python
kmeans=KMeans(3)
kmeans.fit(x)
y_cluster=kmeans.fit_predict(x)
y_cluster
```

```python
data_kmeans=data.copy()
data_kmeans['pred_cluster']=y_cluster
data_kmeans
```

```python
silhouette_avg = silhouette_score(x, y_cluster)
print(f"Silhouette Score: {silhouette_avg}")
```

```python
data=data.drop(columns=['CUST_ID'], axis=1)
```

```python
from sklearn.preprocessing import normalize
data_scaled=normalize(data)
data_scaled=pd.DataFrame(data_scaled, columns=data.columns)
data_scaled
```

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(10, 7))
plt.title('Dendrogram')
dend=sch.dendrogram(sch.linkage(data_scaled, method='ward'))
plt.axhline(y=23, color='r', linestyle='--')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
y_predict=cluster.fit_predict(data_scaled)
y_predict
```

```python
data['pred_cluster']=y_predict
data
```

```python
silhouette_avg = silhouette_score(data_scaled, y_predict)
print(f"Silhouette Score:{silhouette_avg}")
```
